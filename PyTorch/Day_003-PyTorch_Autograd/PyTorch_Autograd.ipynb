# Custom Gradient

```python
# y = x^2
def dx_dy(x: int):
  return 2*x

dx_dy(2) # 2x
```

# Using `PyTorch Grad`

```python
# Make X
x = torch.tensor(2.0, requires_grad=True)
print(f'x = {x}')
display(x)

# Calculate Y
y = x**2
print(f'\ny = {y}')
display(y)

# Calculate Gradient
y.backward()

# Print Gradient
print(f'\nPrint Gradient')
x.grad
```

# Make a small neural network

## Input Init

```python
x = torch.rand(size=(3, 2), dtype=torch.float32)
y = torch.randint(size=(3, 1), high=2, low=0, dtype=torch.int32)

display(x)
display(y)
```

## Weight Init

```python
w = torch.rand(size=(1, 2), requires_grad=True)
b = torch.rand(size=(1, 1), requires_grad=True)

w_prev = w.clone()
b_prev = b.clone()

display(w)
display(b)
```

## Training

```python
# z = x @ w.T + b
z = torch.matmul(x, w.T) + b
y_pred = torch.sigmoid(z)
y_pred
```

## Loss Calculate

```python
bce_loss = nn.BCELoss()
loss = bce_loss(y_pred, y.float())
loss
```

## Backpropagation

```python
loss.backward()

print("Backword")
w.grad
```

```python
print(f'Previous weights: {w_prev}')
print(f'Backword Weights: {w.grad}')

print(f'Previous bias: {b_prev}')
print(f'Backword bias: {b.grad}')
```

## Clean Gradient

```python
display(w.grad.zero_())
display(b.grad.zero_())
```

# Vector Gradient

```python
v = torch.tensor([4., 2., 6.,], requires_grad=True)
v
```

```python
y = (v**2).mean()
display(y)

y.backward()

display(v.grad)
```

# Disable Gradient

- option 1 - `requires_grad_(False)`
- option 2 - `detach()`
- option 3 - `torch.no_grad()`

```python
x = torch.rand(size=(1, 4), requires_grad=True)
display(x)

y = x**2
z = torch.sigmoid(y)
zp = z.mean()
display('Loss', zp)

# Backword
zp.backward()
display('Backowrd Value: ', x.grad)
```

## option 1 - `requires_grad_(False)`

```python
x.requires_grad_(False)

# zp.grad
```

```python
y = (x**2).mean()
y.backward() # Give an Error
```

## `detach()`

```python
z = x.detach()
z
```

```python
y = (z**2).mean()
y.backward() # Give an Error
```

## `torch.no_grad()`

```python
with torch.no_grad():
  y = (x**2).mean()
  print(y)
```
