# 1. Arithmetic operation in column

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'a' : [1, 2, 3, 4, 5],
    'b' : [6, 7, 8, 9, 10]
})
```

# 1.1 addition

```python
df['c'] = df['a'] + df['b']
print(f"Adition: \n{df}")
```

#1.2 Subtraction

```python
df['c'] = df['a'] - df['b']
print(f"Subtraction: \n{df}")
```

#1.3 Multiplication

```python
df['c'] = df['a'] * df['b']
print(f"Multiplication: \n{df}")
```

#1.4 division

```python
df['c'] = df['a'] / df['b']
print(f"Division: \n{df}")
```
