#Handling Missing Values

````python
import pandas as pd
````

# 1. import csv file

````python
data = pd.read_csv('6_ex.csv')
print(data.head())
````

````python
d = data.copy()
d = d.loc[:, ['Train no ', 'Name of Trains ', 'Starting station ', 'Off-day ', 'Arrival time ']]

print(d.head())
````

#2. Dropna

# 2.1 Drop by Row

````python

d1 = data.copy()
d1 = d1.loc[:, ['Name of Trains ', 'Starting station ', 'Off-day ', 'Arrival time ']]
print(d1.head(7))
print()

d1.dropna(axis=0, inplace=True)
print(d1.head(7))
````

#2.2 Drop by Column

````python
d2 = d.copy()
print(d2.head(7))
print()

d2.dropna(axis=1, inplace=True)
print(d2.head(7))
````

# 2.3 How parameter

````python
d3 = d.copy()
print(d3.head(7))
print()

d3.dropna(axis=0, how='all', inplace=True) #all vs any
print(d3.head(7))
````

# 2.4 Subset

```
# Delete None data by specific column
```



````python
d4 = d.copy()
print(d4.head(7))
print()

d4.dropna(axis=0, subset=['Off-day '], inplace=True)
print(d4.head(7))
````

# 2.5 thresh parameter



```
# thresh=n, here if any row contain n number of NULL value then it will be remove
```



````python
d5 = d.copy()
print(d5.head(7))
print()

d5.dropna(thresh=3, inplace=True)
print(d5.head(7))
````

# 3. Fillna

# 3.1 Fill all entity

````python
d6 = d.copy()
print(d6.head(7))
print()

d6.fillna(7.17, inplace=True)
print(d6.head(7))
````

  # 3.2 Fill with specific column

````python
d7 = d.copy()
print(d7.head(7))
print()

# d7['Arrival time '].fillna(7.17, inplace=True)
d7 = d7.fillna({'Starting station ' : 'Dhaka ', 'Off-day ' : 'Friday ', 'Arrival time ' : '20:45'})
print(d7.head(7))
````

# 3.3 Fill with forward and backword field


```
# 1. Forward row
# 2. Backword row
# 3. Forward column
# 4. Backword Column
```



````python
d8 = d.copy()
print(d8.head(7))
print()

d8.fillna(method='ffill', inplace=True)
print(d8.head(7))
````

````python
d9 = d.copy()
print(d9.head(7))
print()

d9.fillna(method='bfill', inplace=True)
print(d9.head(7))
````

````python
d10 = d.copy()
print(d10.head(7))
print()

d10.fillna(method='ffill', axis=1, inplace=True)
print(d10.head(7))
````

````python
d11 = d.copy()
print(d11.head(7))
print()

d11.fillna(method='bfill', axis=1, inplace=True)
print(d11.head(7))
````

# 3.4 Limit method


```
# When i should fill limited data of a row, then use it
```



````python
d12 = d.copy()
print(d12.head(7))
print()

d12.fillna('ABCD', limit=1, inplace=True)
print(d12.head(7))
````

# 4. Replace


#4.1 replace_to 7 to 9

````python
d13 = d.copy()
print(d13.head(7))
print()

d13.replace(to_replace='Dhaka ' , value='Rajshahi', inplace=True)
print(d13.head(7))
````

# 4.2 replace using dictorionary

````python
d14 = d.copy()
print(d14.head(7))
print()

d14.replace(to_replace={'Dhaka ' : 'Rajshahi', 'Friday ' : 'Saturday'}, inplace=True)
print(d14.head(7))
````

# 4.3 Regex

````python
d15 = d.copy()
print(d15.head(7))
print()

d15.replace(to_replace='[A-Za-z]', value='0', regex=True, inplace=True)
print(d15.head(7))
````

# 4.4 method, limit, inplace same as fillna

````python
d16 = d.copy()
print(d16.head(7))
print()

d16.replace('Friday ', method='ffill', limit=1, inplace=True)
print(d16.head(7))
````

# 5.1 Noraml Interpolate

````python
dx = pd.read_csv('student_data.csv')
d17 = dx.copy()
print(d17.head(7))
print()

d17.interpolate(method='linear', inplace=True)
print(d17.head(7))
````

# 5.2 limit-Direction


```
# backword
# forward
#both
```



````python
d18 = dx.copy()
print(d18.head(7))
print()

d18.interpolate(method='linear', limit=1, limit_direction='backward', inplace=True)
print(d18.head(7))
````

# 5.3 Limit area


```
inside
outside
```



````python
d19 = dx.copy()
print(d19.head(7))
print()

d19.interpolate(limit_area='outside', inplace=True)
print(d19.head(7))
````
