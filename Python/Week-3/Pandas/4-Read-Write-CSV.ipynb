#Read Write CSV

```python
import pandas as pd;
```

# 1. Create CSV

```python
dict = {
    'a' : [1,2,3,4,5],
    'b' : [6,7,8,9,10],
    'c' : [11,12,13,14,15],
    'd' : [16,17,18,19,20],
    'e' : [21,22,23,24,25],
}
df = pd.DataFrame(dict)
print(df)
```

```python
df.to_csv('text0.csv')
```

```python
df.T.to_csv('text1.csv', index=False)
```

```python
df.to_csv('text2.csv', index=False, header=['v', 'w', 'x', 'y', 'z']) #header=False or header=['x', 'y']
```

# 2. Read CSV with various parameter

```python
data = pd.read_csv('student_data.csv')
print(data)
```

#2.1 header, rows, usecols

```python
data2 = pd.read_csv('student_data.csv', nrows=5, usecols=[1, 4])
print(data2)
```

```python
data3 = pd.read_csv('student_data.csv', nrows=5, usecols=["ID", "CGPA"])
print(data3)
```

```python
data4 = pd.read_csv('student_data.csv', nrows=5, header=None)
print(data4)
print()

#skip wors
data5 = pd.read_csv('student_data.csv', nrows=5, skiprows=[0, 1])
print(data5)
```

# 2.2 index change into colum

```python

data6 = pd.read_csv('student_data.csv', nrows=5, index_col="ID")
print(data6)
print()

#Header changes into another rows
data7 = pd.read_csv('student_data.csv', nrows=5, header=1)
print(data7)
```

# 2.3 Pesonal Header using prefix, header, names

```python
data8 = pd.read_csv('student_data.csv', nrows=5, header=None, skiprows=1)
print(data8)
```

```python
data9 = pd.read_csv('student_data.csv', nrows=5, header=None, skiprows=1, names=['a', 'b', 'c', 'd', 'e'])
print(data9)
```

```python
data10 = pd.read_csv('student_data.csv', nrows=5, header=None)
print(data10)
```
