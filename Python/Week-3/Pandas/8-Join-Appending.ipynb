# Join and Appending

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
})
df2 = pd.DataFrame({
    'C': ['C0', 'C1', 'C2',],
    'D': ['D0', 'D1', 'D2',],
    'E': ['E0', 'E1', 'E2',],
})
```

#1. Join

#1.1 Usual join

```python
p1 = df1.copy()
p2 = df2.copy()
print(f'Before Join \n{p1}')
print(f'Before Join \n{p2}')

p1 = p1.join(p2, how='outer', lsuffix='_1', rsuffix='_2')
print(f'\nAfter Join \n{p1}')
```

# 2. Append

```python
p3 = df1.copy()
p4 = df2.copy()
print(f'Before Append \n{p3}')
print(f'Before Append \n{p4}')
print()

x=p3._append(p4)
print(f'After Append within index\n{x}')

y=p3._append(p4, ignore_index=True)
print(f'After Append with ignore index\n{y}')
```
