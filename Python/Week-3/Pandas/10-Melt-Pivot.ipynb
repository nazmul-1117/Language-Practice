#Pivot and Melt Function

```python
import pandas as pd
```

```python
data = pd.DataFrame(
    {
        'days' : [1, 2, 3,4,5],
        'name' : ['a','b','c','d','e'],
        'algo' : [26,17,25,10,12],
        'web' : [15,25,23,21,12],
        'micro' : [12, 25, 21, 15, 19],
    }
)
data
```

#1. Melt function

```python
# Convert the data into long format

d1 = data.copy()
print(pd.melt(d1))
```

#1.1 change name of column

```python
d2=data.copy();
d2 = pd.melt(d2, id_vars=['algo'], var_name='Course', value_name='Marks')
print(d2)
```

# 2. Pivot function

```python
p1 = data.copy()
print(p1)
print()

x = p1.pivot(index='days', columns='name')

print(x)
```

# 2.1 get single data

```python
p2 = data.copy()
print(p2)
print()

x = p2.pivot(index='days', columns='name', values='algo')
print(x)
print()

x = p2.pivot(index='days', columns='name', values=['web', 'micro'])
print(x)
```

```python
data2 = pd.DataFrame(
    {
        'days' : [1, 2, 1,1,2],
        'name' : ['a','b','a','b','a'],
        'algo' : [26,17,25,10,12],
        'web' : [15,25,23,21,12],
        'micro' : [12, 25, 21, 15, 19],
    }
)

# aggfunc --> mean,mediun,mode,sum,count,min,max

data2.pivot_table(index='days', columns='name', aggfunc='sum', margins='True')
```
