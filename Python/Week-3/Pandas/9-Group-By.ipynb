```python
import pandas as pd
```

```python
data = pd.DataFrame(
    {
        'Name' : ['Nazmul', 'Fuad', 'Nazmul', 'Sojib', 'Fuad', 'Nazmul', 'Sojib', 'Fuad',
                  'Sojib', 'Nazmul', 'Fuad', 'Sojib', 'Nazmul', 'Fuad', 'Sojib'],
        'cgpa1' : [3.52, 3.2, 3.52, 3.7, 2.6, 3.5, 3.2, 2.9, 3.7, 2.6, 3.5, 3.2, 2.9, 3.7, 2.6],
        'cgpa2' : [3.37, 3.42, 3.49, 2.92, 3.88, 3.27, 2.35, 2.78, 3.76, 3.23, 2.2, 2.09, 2.84, 2.37, 3.29],
        'cgpa3' : [3.52, 3.56, 2.39, 3.26, 3.96, 2.58, 3.84, 2.99, 2.24, 3.92, 2.13, 2.65, 2.79, 2.83, 2.26]
    }
)

data
```

# 1. group_by

```python
print(data)
```

```python
d1=data.copy();

c = d1.groupby('Name')

print(c)
for x, y in c:
  print(x)
  print(y)
  print()
```

#1.1 Group by by multiple column

```python
x = data.groupby(['Name', 'cgpa1'])
print(x)
print(x.first())

# for i, j in x:
#   print(i)
#   print(j)
#   print()

```
