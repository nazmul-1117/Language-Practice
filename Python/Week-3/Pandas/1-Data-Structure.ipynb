# Data structure in pandas

# 1. Series

```python
list1 = [1,2,3,4,5]
list2 = [7.5, 6.5, 8.5, 17.2, 8.9,]
list3 = ['a', 'b', 'c', 'd', 'e']
```

```python
dict1 = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5}
dict2 = {'a':7.5, 'b':6.5, 'c':8.5, 'd':17.2, 'e':8.9}
dict3 = {'a':'a', 'b':'b', 'c':'c', 'd':'d', 'e':'e'}
```

```python
tuple1 = (1,2,3,4,5)
tuple2 = (7.5, 6.5, 8.5, 17.2, 8.9,)
tuple3 = ('a', 'b', 'c', 'd', 'e')
```

# 1.1 Series Create

```python
# Simple Series with list
l1 = list1.copy()
s1 = pd.Series(l1)
print(s1)
print(type(s1))
```

```python
# changing index
s2 = list2.copy()
s2 = pd.Series(s2, index = list3, name="Firts_Pandas_File") #list3 = [a, b, c, d, e]
print(s2)
print(s2.dtype)

```

```python

s3 = pd.Series(tuple1, index = list3, name="Firts_Pandas_File") #list3 = [a, b, c, d, e]
print(s3)
print(s3.dtype)
```

```python
s4 = pd.Series(dict1)
print(s4)
```

#2. Data Frame

```python
d = {
    'id' : [80, 88, 89, 158, -1],
    'name' : ["Fuad", "Sojib", "Nazmul", "Araf", "X"],
    'age' : [21, 22.5, 22.6, 23, 999.99],
    'cgpa' : [4.5, 4.6, 4.7, 4.8, 4.9]
}

lx = ['Fuad', 'Sojib', 'Nazmul', 'Araf', 'X']
ly = [80, 88, 89, 158, -1]
lz = [4.5, 4.6, 4.7, 4.8, 3.9]
```

```python
d1 = d.copy();
df1 = pd.DataFrame(d1)
print(df1)
```

```python
# play with index
d2 = d.copy();
i = [1, 2, 3, 4, 5]
df2 = pd.DataFrame(d2, columns = ['id', 'name'], index =i)
print(df2)
```

# 2.1 Play with data

```python
d3 = d.copy();
df3 = pd.DataFrame(d3)
print(df3)
print()

print(f"Row 3, Column 2 [Nazmul]: {df3['name'][2] }")
print(f"Row 2, Column 2: {df3.iloc[3,3]}")
```

# 2.2 List to DataFrame

```python
l1 = list1.copy()
l2 = list2.copy()
l3 = list3.copy()
df4 = pd.DataFrame([l1, l2, l3])
print(df4)
```

# 2.3 Series to Data Frame

```python
l1 = lx.copy()
l2 = ly.copy()
l3 = lz.copy()
print(f"List 1: {l1}")
print(f"List 2: {l2}")
print(f"List 3: {l3}")
print()

s1 = pd.Series(l1)
s2 = pd.Series(l2)
s3 = pd.Series(l3)
print(f"Series 1: \n{s1}")
print(f"\nSeries 2: \n{s2}")
print(f"\nSeries 3: \n{s3}")
print()

df5 = pd.DataFrame([s1, s2, s3], index = ['Name', 'ID', 'CGPA'])
print(df5)

# s1 = pd.Series(l1)
# df5 = pd.DataFrame(s1)
# print(df5)
```
