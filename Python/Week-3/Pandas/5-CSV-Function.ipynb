# 5. Some Drop Method

```python

z = data.copy()
z = z.loc[:, ['Train no ', 'Name of Trains ', 'Starting station ', 'Arrival Station']]
print(z.head())
print()

# Drop row
z.drop(index=0, inplace=False)
print(z.head())
print()

# drop column
z.drop(columns=['Starting station '], inplace=False)
print(z.head())
print()
```

```python
# Drop another way
z = data.copy()
z = z.loc[:, ['Train no ', 'Name of Trains ', 'Starting station ', 'Arrival Station']]
print(z.head())
print()

# Drop row
z.drop(0, axis=0, inplace=True)
print(z.head())
print()

# drop column
z.drop('Starting station ', axis=1, inplace=True)
print(z.head())
print()
```
