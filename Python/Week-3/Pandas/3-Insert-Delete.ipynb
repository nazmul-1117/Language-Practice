# Insert, Delete on Column

```python
import pandas as pd
```

```python
d = {'a': [1, 2], 'b': [3, 4]}
df = pd.DataFrame(d)
```

# 1. Insert

```python
df1 = df.copy()
print(f'Original Data: \n{df1}')
print(type(df1))
print()

# insert(index, name, data)
df1.insert(2, 'c', [5, 6])
print(f'New Data: \n{df1}')
print(type(df1))
```

```python
# Copy Column
df2 = df.copy()
print(f'Original Data: \n{df2}')
print(type(df2))
print()

# copy()
# df2['d'] = df2['a']
df2.insert(2, 'c', df2['b'])
print(f'New Data: \n{df2}')
print(type(df2))
```

```python
dfx = df.copy()
print(f'Original Data: \n{dfx}')
print(type(dfx))
print()

dfx = dfx.assign(d = dfx['a'])
print(f'New Data: \n{dfx}')
print(type(dfx))
```

```python
dfy = df.copy()
print(f'Original Data: \n{dfy}')
print(type(dfy))
print()

dfy['d'] = dfy['a'][:1]
dfy.loc[:, 'e'] = [7, 8]
print(f'New Data: \n{dfy}')
```

# 2. Delete

```python
df3 = df.copy()
print(f'Original Data: \n{df3}')
print(type(df3))
print()

# delete(name)
del df3['a']
print(f'New Data: \n{df3}')
```

```python
#Another way
df4 = df.copy()
print(f'Original Data: \n{df4}')
print(type(df4))
print()

# pop(name)
dl = df4.pop('b')
print(f'New Data: \n{df4}')

print(f'Deleted Data: \n{dl}')
```
