# Drop Column above 20% (standard)



```
Here some step to delete data
1. Find the percentage of null value in each column
2. Get that column name whose null value greater than 20
3. Delete that column 'list' from main Data frame
```



````python
# Step1: show all percentage data
drop_column = df.copy()
drop_percentage_column = drop_column.isnull().sum()*100/df.shape[0]

drop_percentage_column
````

````python
# show data of greater than 0
drop_percentage_column[drop_percentage_column > 0]
````

````python
# Get the column list of those column
x = drop_percentage_column[drop_percentage_column > 17]
x.keys()
````

# Delete Column

````python
# Drop that list from main data frame
print(f'Before column drop: {df.shape}')
drop_col = df.drop(columns=x.keys())
print(f'After column drop: {drop_col.shape}')
````

````python
# After drop 20% above null values column
sns.heatmap(drop_col.isnull())
````

# Delete Rows with Null value

````python
print(f'Before row drop: {drop_col.shape}')
drop_col_row = drop_col.dropna()
print(f'After row drop: {drop_col_row.shape}')
````
