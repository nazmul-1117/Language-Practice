# Some Special array

```python
import numpy as np
```

# 1. Array filled with 0

```python
arr1 = np.zeros((3, 3))
print(arr1)
```

# 2. Array filled with 1

```python
arr2 = np.ones((3, 3))
print(arr2)
```

# 3. create empty array

```python
arr3 = np.empty((2, 3, 3))
print(arr3)
```

# 4. Array with range

```python
arr4 = np.arange(1, 11)
print(arr4)
```

# 5. Array with diagonal elements 1

```python
arr5 = np.eye(4)
print(arr5)
```

# 6. Create with value with specific interval

```python
arr6 = np.linspace(0, 20, num=5);
print(arr6)
```
