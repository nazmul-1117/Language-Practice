# Matrix

```python
import numpy as np
```

```python
arr1 = np.array([
        [5,6,7],
        [8,9,10],
        [11,12,13]
    ])

arr2 = np.array([
        [1,2,3],
        [3,2,1],
        [3,2,1]
    ])
```

```python
mtx1 = np.matrix([
        [5,6,7],
        [8,9,10],
        [11,12,13]
    ])

mtx2 = np.matrix([
        [1,2,3],
        [3,2,1],
        [3,2,1]
    ])
```

#difference between array and matrix

#1. Matrix

```python
a =mtx1.copy()
b = mtx2.copy()

print(f"Matrix A: \n{a}")
print(f"Matrix B: \n{b}")
print()

# Matrix addition
c = a + b
print(f"Matrix Addition: \n{c}")
print()

# Matrix subtraction
d = a - b
print(f"Matrix Subtraction: \n{d}")
print()

# Matrix dot productu


e = a * b
print(f"Matrix Dot Product: \n{e}")
print()

print(f"Type {type(c)}, Dimension: {  d.shape  }")
print(f"Type {type(e)}, Dimension: {  e.shape  }")
```

```python
a = arr1.copy()
b = arr2.copy()

print(f"Array A: \n{a}")
print(f"Array B: \n{b}")
print()

# Matrix addition
c = a + b
print(f"Array Addition: \n{c}")
print()

# Matrix subtraction
d = a - b
print(f"Array Subtraction: \n{d}")
print()

# matrix dot product
e = a * b
print(f"Array Dot Product: \n{e}")
print()

print(f"Type {type(c)}, Dimension: {  d.shape  }")
print(f"Type {type(e)}, Dimension: {  e.shape  }")
```

```python
x = mtx1.copy()
y = mtx2.copy()
y = np.delete(y, 1, axis=1)


print(f"Original Array A: \n{x}")
print(f"\nOriginal Array B: \n{y}")

print(f"Dimension A: {x.shape}")
print(f"Dimension B: {y.shape}")
print()

# Matrix Multiplication
z = np.dot(x, y)
print(f"Matrix Multiplication: \n{z}")
print()
```

# 2. Transpose

```python
x = np.matrix([
    [11,12,13],
    [21, 22, 23],
])
print(f"Original Array: \n{x}")
print(f"Dimension: {x.shape}")

# transpose
x = np.transpose(x)
print(f"\nTranspose Array: \n{x}")
print(f"Dimension: {x.shape}")

# Another way
x = x.T
print(f"\nTranspose Array: \n{x}")
print(f"Dimension: {x.shape}")
```

# 3. Swap axes

```python
x = np.matrix([
    [11,12,13],
    [21, 22, 23],
])
print(f"Original Array: \n{x}")
print(f"Dimension: {x.shape}")
print()

x = np.swapaxes(x, 0, 1)
print(f"Swap Axes Array: \n{x}")
print(f"Dimension: {x.shape}")
```

# 4. Inverse

```python
x = np.matrix([
    [1,2],
    [3,4]
])
print(f"Original Array: \n{x}")
print(f"Dimension: {x.shape}")
print()

x = np.linalg.inv(x)
print(f"Inverse Array: \n{x}")
```

#5. Power

A<sup>n</sup> <br/>
Case 1:  n > 0 --> A has Normal Power <br/>
Case 2: n==0 --> A is Identity matrix<br/>
Case 3: n < 0 --> A has power and multiply with A's Inverse matrix<br/>

```python


x = np.matrix([
    [1,2],
    [3,4]
])
print(f"Original Array: \n{x}")
print(f"Dimension: {x.shape}")
print()

k = np.linalg.matrix_power(x, 3)
print(f"Power of 3: \n{k}")
print()

l = np.linalg.matrix_power(x, -1)
print(f"Power of -1: \n{l}")
print()

m = np.linalg.matrix_power(x, 0)
print(f"Power of 0: \n{m}")
print()
```

# 6. Determinate of matrix

```python
x = np.matrix([
    [1,2,],
    [3,4,]
])
print(f"Original Array: \n{x}")
print(f"Dimension: {x.shape}")
print()

k = np.linalg.det(x)
print(f"Determinate: \n{k}")
print()
```
