# Random Number Generate and filled with matris

```python
#Random Number
print(np.random.rand()) #0 -> 1
print(np.random.randn())  # negggave -> 0
print(np.random.ranf()) #0.0 -> 0.99n9
print(np.random.randint(1, 10)) # int between min -> max
```

```python
import numpy as np
```

# 1. Rand() --> Number between 0-1

```python
mtx1 = np.random.rand(3, 3)
print(mtx1)
```

# 2. Randn --> negative and close to zero

```python
mtx2 = np.random.randn(3, 3)
print(mtx2)
```

# 3. Ranf() --> 0.00 - 0.9n9

```python
mtx3 = np.random.ranf((3, 3))
print(mtx3)
```
