# Arithmetic Function

```python
import numpy as np
```

# 1. Shuffle

```python
mtx1 = np.arange(10)
print(f"Original Matrix: {mtx1}")

np.random.shuffle(mtx1)
print(f"Shuffled Matrix: {mtx1}")
```

```python
mtx2 = np.arange(9).reshape(3,3)
print(f"Original Matrix: \n{mtx2}")

np.random.shuffle(mtx2)
print(f"\nShuffled Matrix: \n{mtx2}")
```

# 2. Unique

```python
l = [1,2,2,3,3,3]
mtx3 = np.array(l)

print(f"Original Matrix: {mtx3}")

print(f"\nUnique Matrix: {np.unique(mtx3)}")
```

```python
mtx4 = np.array([
    [1, 2, 2],
    [3, 1, 5],
    [5, 1, 2],
],)
print(f"Original Matrix: \n{mtx4}")

print(f"\nUnique Matrix: {np.unique(mtx4)}")
```

# 3. Resize

```python
#Resize array and i can convert this to 2-D
mtx5 = np.array([1,2,3,4, 5, 6, 7])
print(f"Original Matrix: {mtx5}")

x = np.resize(mtx5, (4,3))
print(f"\nResized Matrix: \n{x}")
print(f"\nShape: {x.shape}")
```

#4. Flatten

order:
  1. C : c-style --> row wise
  2. F : Fortan --> column wise
  3. A :
  4. K

```python
# COnvert 2-D array to 1-D array
mtx6 = np.array([
    [1, 2, 2],
    [3, 1, 5],
    [5, 1, 2],
],)
print(f"Original Matrix: \n{mtx6}")

x = mtx6.flatten()
print(f"\nFlatten Matrix: {x}")

x[0] = 100
print(f"\nAfter Change values of 0's index")
print(f"Original Matrix: \n{mtx6}")
print(f"Flatten Matrix: {x}")
print()

```

```python
# Play with order
print(f"Original Matrix: \n{mtx6}")

v = mtx6.flatten(order='C')
print(f"\nFlatten Matrix - C: {v}")

x = mtx6.flatten(order='F')
print(f"\nFlatten Matrix - F: {x}")

y = mtx6.flatten(order='A')
print(f"\nFlatten Matrix - A: {y}")

z = mtx6.flatten(order='K')
print(f"\nFlatten Matrix - K: {z}")
```

#5. Ravel

```python
# COnvert 2-D array to 1-D array
mtx7 = np.array([
    [1, 2, 2],
    [3, 1, 5],
    [5, 1, 2],
],)
print(f"Original Matrix: \n{mtx7}")

x = np.ravel(mtx7)
print(f"\nFlatten Matrix: {x}")

x[0] = 100 #Ravel is same like pass by reference, change original value
print(f"\nAfter Change values of 0's index")
print(f"Original Matrix: \n{mtx7}")
print(f"Flatten Matrix: {x}")
```

```python
# Order practice
print(f"Original Matrix: \n{mtx7}")

z = np.ravel(mtx7, order='C')
print(f"\nFlatten Matrix - C: {z}")

x = np.ravel(mtx7, order='F')
print(f"\nFlatten Matrix - F: {x}")

t = np.ravel(mtx7, order='A')
print(f"\nFlatten Matrix - A: {t}")

y = np.ravel(mtx7, order='K')
print(f"\nFlatten Matrix - K: {y}")
#
```
