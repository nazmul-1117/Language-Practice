# Data type in numpy same as c/c++

```python
import numpy as np
```

```python
mtx1 = np.array([1,2,3,4,5,6], dtype=np.int8)
print(mtx1)
print(type(mtx1))
print(mtx1.dtype)
```

# conversion of between data type

```python
mtx2 = np.array([1,2,3,4,5,6], dtype=np.int32)
print(mtx2)
print(type(mtx2))
print(mtx2.dtype)

print("\nConvert INT to FLOAT")
mtx3 = mtx2.astype(np.float32)
print(mtx3)
print(type(mtx3))
print(mtx3.dtype)

print("\nConvert FLOAT to INT")
mtx4 = np.int32(mtx3)
print(mtx4)
print(type(mtx4))
print(mtx4.dtype)
```
