# 9 Hidden Features of Python

# 1. `else` with `for` loop
- The else block runs only if the loop finish without a break

```python
for i in range(5):
  print(i)
else:
  print('For loop Done without break statement')
```

```python
for i in range(5):
  if i == 3:
    break
  print(i)
else:
  print('For loop Done without break statement')
```

# 3. Multiple Assignment
- Python allows multiple assignment in one line

```python
x, y, z = 1, 2, 3
print(x, y, z)
```

# 4. `zip()` method for parallel iteration
- iterate over multiple lists in parallel at once

```python
names = ['alice', 'bob', 'don', 'omar', 'johny']
ages = [15, 30, 35, 10, 52]
countries = ['usa', 'uk', 'rus', 'uae', 'aus',]

for name, age, country in zip(names, ages, countries):
  print(name, age, country)
```

# 5. `enumerate()` method for Track Index
- for manual counter it helps more

```python
names = ['alice', 'bob', 'don', 'omar', 'johny']
for i, name in enumerate(names):
  print(f'index: {i}, name: {name}')
```

# 7. Swap two variables without temp
- swap two variables in one line

```python
x = 10
y = 20
x, y = y, x
print(x, y)
```

# 8. `else` with `try/execpt` block
- the `else` block runs only when there is no error occur in `try/except` block

```python
try:
  print("Hi I am try block")
except:
  print("Hi I am except block")
else:
  print("Hi I am else block")
```

```python
try:
  print("Hi I am try block")
  print(1/0)
except:
  print("Hi I am except block")
else:
  print("Hi I am else block")
```

# 9. if `__name__ == "__main__"`
- Lets control when program runs

```python
print("I am top of the program")

def add(a, b):
  print(a+b)

if __name__ == "__main__":
  add(10, 20)
```
