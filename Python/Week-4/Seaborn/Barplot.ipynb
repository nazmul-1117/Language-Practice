# Barplot in Seaborn

# Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#import dataset

```python
df = pd.read_csv('Iris.csv')
```

```python
df.shape
```

# Barplot

```python
plt.figure(figsize=(15,5))
sns.barplot(data = df, x='Species', y='SepalLengthCm')
```

#Hue

```python
plt.figure(figsize=(15,5))
sns.barplot(data = df, x='Species', y='SepalWidthCm', hue='Species')
```

# Palette

```python
plt.figure(figsize=(15,5))
sns.barplot(data = df, x='Species', y='SepalWidthCm', hue='Species', palette='rainbow')
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data = df, x='Species', y='SepalWidthCm', hue='Species', palette=['red','blue','green'])
```

```python
sns.barplot(data = df, x='Species', y='SepalWidthCm', hue='Species', palette='dark:green')
```

# Color

```python
sns.barplot(data = df, x='Species', y='SepalWidthCm', hue='Species', color='green')
```
