# Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load dataset

```python
df = pd.read_csv('Iris.csv')
df_b = df.copy()
```

#Line Plot

```python
plt.figure(figsize=(15,3))
sns.lineplot(x='SepalWidthCm', y='SepalLengthCm', data=df)
```

# 1. Hue parameter

```python
plt.figure(figsize=(15,4))
sns.lineplot(x='PetalWidthCm', y='PetalLengthCm', data=df, hue='Species')
```

# 2. stylename parameter

```python
plt.figure(figsize=(15,4))
sns.lineplot(x='PetalWidthCm', y='PetalLengthCm', data=df, hue='Species', style='Species')
```

# 3. markers

```python
plt.figure(figsize=(15,5))
sns.lineplot(x='SepalLengthCm', y='SepalWidthCm', data=df, hue='Species', style='Species', markers=True)
```

# 4. Palette

```python
plt.figure(figsize=(15,5))
sns.lineplot(
    data=df,
    x='SepalLengthCm',
    y='SepalWidthCm',
    hue='Species',
    style='Species',
    markers=True,
    palette='dark'
    )
```
