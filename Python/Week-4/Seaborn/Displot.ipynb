# Displot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
df.shape
```

# Displot

```python
plt.figure(figsize=(15,5))
sns.displot(data=df, x='Age')
```

# Kind = 'kde/hist'

```python
plt.figure(figsize=(15,5))
sns.displot(data=df, x='Age', kind='kde')
```

# Hue

```python
plt.figure(figsize=(15,5))
sns.displot(data=df, x='Age', hue='Sex')
```

# Palette

```python
plt.figure(figsize=(15,5))
sns.displot(data=df, x='Age', hue='Sex', palette='rocket')
```
